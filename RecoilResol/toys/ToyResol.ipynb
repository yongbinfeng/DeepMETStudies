## Notebook to check resolution effects

```python
import ROOT
```

```python
rand = ROOT.TRandom3()
```

```python
h1 = ROOT.TH1F("h1", "h1_GenMET", 100, 100, 200)
h2 = ROOT.TH1F("h2", "h2_RecoMET", 100, 100, 200)
h3 = ROOT.TH1F("h3", "h3_METDiff", 200, -100, 100)
h4 = ROOT.TH1F("h4", "h4_halfMETDiff", 400, -300, 100)
h5 = ROOT.TH1F("h5", "h5_METRatio", 200, 0.5, 1.5)
```

```python
for i in range(100000):
    gMET = ROOT.TMath.Exp(rand.Uniform(0, 10))
    rMET = rand.Gaus(gMET, 20.0) * 1.5
    h1.Fill(gMET)
    h2.Fill(rMET)
    h3.Fill(rMET - gMET)
    h4.Fill(rMET / 1.5 - gMET)
    h5.Fill(rMET / gMET)
```

```python
c1 = ROOT.TCanvas("c1", "c1", 800, 600)
h1.Draw()
c1.Draw()
```

```python
h3.Draw()
print(h3.GetRMS())
c1.Draw()
```

```python
h4.Draw()
print(h4.GetRMS())
c1.Draw()
```
